## 用读取的数据做出交叉表

```python
ddd=pd.crosstab(df.app_name, df.kind_comment, margins=True)
ddd
```

```python
ddd1=pd.crosstab(df.app_name, df.kind_comment)
ddd1
```

## 计算出好评率

```python
rate_good=pd.DataFrame(ddd['好评']/ddd['All'],columns=['好评率'])
rate_bad=pd.DataFrame(ddd['差评']/ddd['All'],columns=['差评率'])
rate_medium=pd.DataFrame(ddd['中评']/ddd['All'],columns=['中评率'])
```

```python
print(rate_good)
print(rate_bad)
print(rate_medium)
```

```python
index_name=rate_good.index
```

```python
index_name
```

## 数据可视化

```python
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
# 创建一个点数为 8 x 6 的窗口, 并设置分辨率为 80像素/每英寸
plt.figure(figsize=(60, 40), dpi=400)

# 再创建一个规格为 1 x 1 的子图
plt.subplot(1, 1, 1)

# 柱子总数
N = len(index_name)
# 包含每个柱子对应值的序列
bad = rate_bad['差评率']
good=rate_good['好评率']
medium=rate_medium['中评率']


# 包含每个柱子下标的序列
index = np.arange(N)

# 柱子的宽度
width = 0.35

# 绘制柱状图, 每根柱子的颜色为紫罗兰色
p2 = plt.bar(index, bad, width=0.6, label="差评率", color="r")
p3=plt.bar(index, good, width=0.6, label="好评率", color="#87CEFA",bottom=bad )
p4=plt.bar(index, medium, width=0.6, label="中评率", color="#98FB98",bottom=bad +good )
for x,y,z in zip(index, bad+good+medium,good):
    plt.text(x+0.05,y+0.01,'%.3f' % z,ha='right',va='top') # 使用matplotlib画柱状图并标记数字
# 设置横轴标签
plt.xlabel('App',size=50)
# 设置纵轴标签
plt.ylabel('百分比',size=40)

# 添加标题
plt.title('360应用商店app好评度',loc='left',size=15)

# 添加纵横轴的刻度
plt.xticks(index, index_name,rotation=90)

plt.tick_params(labelsize=23)


plt.yticks(np.arange(0, 1.1, 0.1))

# 添加图例
plt.legend(loc="upper right", fontsize=40)
plt.savefig(r'C:\Users\hayderwang\Desktop\2.png')
plt.show()
```

## 导入词云需要的包

```python
from wordcloud import WordCloud
import PIL.Image as image
image=plt.imread('1.jpg')
```

## 生成词云

```python
df3=df.groupby('app_name')
df2=df.groupby('app_name').groups
i=1
image=plt.imread('1.jpg')
for name,group in df3:
    print (name)
    # print (group)
    print('*'*100)
    content=group['comment']
    c=''.join(content).strip()
    print(c)
    wc = WordCloud(
        font_path=r'D:\python\py3.7\Lib\site-packages\wordcloud\simhei.ttf',
        background_color='white',
        max_words=1000,
        mask=image,
        scale=8,
        margin=30,
        max_font_size=150,  # 设置字体最大值
    ).generate(c)
    # image_produce = WordCloud.to_image()
    # image_produce.show()
    wc.to_file(name+'.png')
    print(i)
    i+=1
    print('*'*100)
```
